# JupyterHub
Multi-Tenant notebook implementation
Implementation steps

First we set the name of the server to install on. We also define the an alias 'remote' to ease writing commands ot the server

````bash
export SERVER=kact-srv-001.kact.sblokalnet
export IPA_SERVER=kact-adm-001.kact.sblokalnet
export LUSER=abr-sadm

function remote(){
ssh $LUSER@$SERVER $@
}
export remote

function remotesudo(){
#Because sudo gives stupid errors if you do not run it from a location that root can see 
#root cannot see automounted home dirs)
ssh $LUSER@$SERVER cd /tmp '&&' sudo "bash -l -c '$@'"
}
export remotesudo

function ipa(){
ssh $LUSER@$IPA_SERVER ipa $@
}
export ipa

function remoteadmsudo(){
#Because sudo gives stupid errors if you do not run it from a location that root can see 
#root cannot see automounted home dirs)
ssh $LUSER@$IPA_SERVER cd /tmp '&&' sudo "bash -l -c '$@'"
}
export remoteadmsudo
````

### Jupyterhub Sudo Spawner

To make jupyterhub run each notebook as the logged in user, install the sudospawner plugin

````bash
remotesudo pip3 install -U -I git+https://github.com/jupyter/sudospawner
````

To make sudospawner work with kerberos, a specific executable have to be installed

````bash
#Find sudospawner 
sudoSpawnerExec=$(remote whereis -b sudospawner | cut -d':' -f2)

#This is where we will put our replacement
newSudoSpawner=$(dirname $sudoSpawnerExec)/sudospawner-singleuser
echo $newSudoSpawner

python << EOJ | remotesudo tee $newSudoSpawner
print("""#!/bin/bash
set -e
#This is executed as the logged in user

#The KerberosAuthenticator (see jupyterhub_config.sh) sets the env KERBEROS_PASSWORD

# Use kinit to do authentication again, this time after sudoing to the desired notebook
# server user so that the kerberos cred cache is created
echo \$KERBEROS_PASSWORD | kinit -V

unset KERBEROS_PASSWORD
# Clear the kerberos password that we set in the environment in our custom spawner so that
# its not visible to the notebook server and kernels.

# Delegate launch to the notebook server.
exec "\$(dirname "\$0")/jupyterhub-singleuser" \$@
""")
EOJ

remotesudo chmod a+x $newSudoSpawner
````

##### Generate a SSL certificate
Skip this step if you already have a ssl certificate you want to use

From http://stackoverflow.com/a/26093147/4527948

````bash
RNDPASS=jupyterHubTestPass

remote RNDPASS=$RNDPASS \
    openssl genrsa -out jupyterhub.key -passout pass:$RNDPASS -aes256 4096
````

````bash
remote "RNDPASS=$RNDPASS echo $RNDPASS > jupyterhub.key.pass"
````

````bash
remote RNDPASS=$RNDPASS \
    openssl req -sha256 -passin pass:$RNDPASS -subj "/C=DK/ST=Aarhus/L=Springfield/O=Royal\ Library/CN=$SERVER" -new -key jupyterhub.key -out jupyterhub.csr
````

````bash
remote RNDPASS=$RNDPASS \
    openssl rsa -passin pass:$RNDPASS -in jupyterhub.key -out jupyterhub.cert.key
````

````bash
remote RNDPASS=$RNDPASS \
    openssl x509 -passin pass:$RNDPASS -in jupyterhub.csr -out jupyterhub.cert -req -signkey jupyterhub.cert.key
````

Install the certificate in the expected location

````bash
remote cp jupyterhub.cert /tmp/
remotesudo mv /tmp/jupyterhub.cert /etc/ssl/certs/jupyterhub.cert
remotesudo chown jupyterhub:jupyterhub /etc/ssl/certs/jupyterhub.cert
remotesudo chmod 0500 /etc/ssl/certs/jupyterhub.cert
````

````bash
remotesudo mkdir -p /etc/ssl/private
remote cp jupyterhub.cert.key /tmp/
remotesudo mv /tmp/jupyterhub.cert.key /etc/ssl/private/jupyterhub.cert.key
remotesudo chown jupyterhub:jupyterhub /etc/ssl/private/jupyterhub.cert.key
remotesudo chmod 0500 /etc/ssl/private/jupyterhub.cert.key
````

### PAM

First we create a separate pam service entry specifically for jupyterhub. This allows us to create FreeIPA HBAC rules later

````bash
remotesudo cp /etc/pam.d/login /etc/pam.d/jupyterhub
````

If you get errors like this in /var/log/messages

    [system] Rejected send message, 2 matched rules; type="method_call", sender=":1.2292" (uid=1031 pid=59107 comm="/opt/anaconda3/bin/python /opt/anaconda3/bin/jupyt") interface="org.freedesktop.login1.Manager" member="CreateSession" error name="(unset)" requested_reply="0" destination="org.freedesktop.login1" (uid=0 pid=823 comm="/usr/lib/systemd/systemd-logind ")

You need to enable the user jupyterhub to handle login sessions. Per default, only processes run as root can do logins. 

````bash
python << EOX | remotesudo tee  /etc/dbus-1/system.d/jupyterhub.conf
print("""
<?xml version="1.0"?> <!--*-nxml-*-->
<!DOCTYPE busconfig PUBLIC "-//freedesktop//DTD D-BUS Bus Configuration 1.0//EN"
        "http://www.freedesktop.org/standards/dbus/1.0/busconfig.dtd">
<busconfig>
        <policy user="jupyterhub">
                <allow send_destination="org.freedesktop.login1"/>
                <allow receive_sender="org.freedesktop.login1"/>
        </policy>
</busconfig>
""")
EOX
````

Restart the dbus deamon. This disconnects the login daemon.... So restart it also

````bash
remotesudo service dbus restart
````

````bash
remotesudo systemctl restart systemd-logind
````

### Systemd Service

Install this systemd service script to make jupyterhub run as a system service

````bash
python << EOS | remotesudo tee  /lib/systemd/system/jupyterhub.service
print("""
[Unit]
Description=Jupyterhub
After=network-online.target

[Service]
#The system user for jupyterhub
User=jupyterhub

#Two ways of specifying the runtime directory
RuntimeDirectory=jupyterhub
WorkingDirectory=/var/run/jupyterhub

#This sets up the logging directory
PermissionsStartOnly=true
ExecStartPre=/usr/bin/mkdir -p /var/log/jupyterhub
ExecStartPre=/usr/bin/touch /var/log/jupyterhub/jupyterhub.log
ExecStartPre=/usr/bin/chown jupyterhub /var/log/jupyterhub -R

#Param explanation
#Config = path to config file
#Log file = the log file we did all the work above to get writable
#ip = the ip where the server will listen. Set to $SERVER, defined in this workbook
ExecStart=/usr/bin/jupyterhub --config=/etc/jupyterhub/jupyterhub_config.py --log-file=/var/log/jupyterhub/jupyterhub.log --ip="kact-srv-001.kact.sblokalnet"

[Install]
WantedBy=multi-user.target
""")
EOS
remotesudo systemctl daemon-reload
````

### Sudo Rules

````bash
ipa sudorule-add jupyterhub --hostcat='all' --runasusercat='all' --runasgroupcat='all'
#Really only need to be able to do it on kact-srv
ipa sudorule-add-user jupyterhub --users=jupyterhub
ipa sudorule-add-option jupyterhub --sudooption='!authenticate'
ipa sudorule-add-option jupyterhub --sudooption='!requiretty'
ipa sudocmdgroup-add 'jupyterhub'
ipa sudocmd-add '/usr/bin/sudospawner'
ipa sudocmdgroup-add-member 'jupyterhub' --sudocmds='/usr/bin/sudospawner'
ipa sudorule-add-allow-command jupyterhub --sudocmdgroups='jupyterhub'
````
